Que.1Explain why we have to use Exception class while creating a custom Exception?

```python
'''When creating a custom exception in Python, it is recommended to inherit from the built-in Exception class or one of its 
subclasses. Here's why using the Exception class is important:

Consistency: Inheriting from the Exception class ensures that your custom exception follows the established hierarchy and 
behavior of the Python exception system. It provides a standardized way to define and handle exceptions.

Compatibility: By inheriting from Exception, your custom exception can be caught and handled along with other built-in
exceptions. It can be used in try-except blocks to handle specific types of exceptions or as a base class for catching 
multiple related exceptions.

Exception Handling: The Exception class defines important methods and attributes that are used for exception handling. 
By inheriting from it, your custom exception can leverage these features, such as __str__() for providing a human-readable
string representation of the exception or args to access the arguments passed to the exception constructor.

Best Practices: Following the established conventions and using the Exception class demonstrates good programming practices. 
It makes your code more readable and understandable for other developers who may encounter your custom exception.

```

Que.2Write a python program to print python exception heirarchy.

```python
import sys
def print_exception_heirarchy(exception_class,indent=0):
    print(''*indent+exception_class.__name__)
    for subclass in exception_class.__subclasses__():
        print_exception_heirarchy(subclass, indent+4)
        
print_exception_heirarchy(Exception)
```

Que.4 Why lookup errors class is used?Explain with an example keyerror and indexerror.

```python
'''The "LookupError" class is a base class in Python that represents errors related to accessing or looking up values in 
a sequence, mapping, or other collection types. It serves as the parent class for more specific lookup-related error
classes, such as "KeyError" and "IndexError."
here are the example of key error and index error
```

```python
my_dict = {'dog': 1, 'cat': 2, 'hamster': 3}

print(my_dict['dog'])  # Output: 1
print(my_dict['bird'])  # Raises KeyError: 'bird' is not a key in my_dict
```

```python
my_list = [1, 2, 3, 4, 5]

print(my_list[2])     # Output: 3
print(my_list[10])    # Raises IndexError: list index out of range
```

Que.6 List down some best practice for exception handling in python.

```python
'''Be specific in exception handling:Catch only the exceptions you expect and can handle. 
Use multiple except blocks: When handling multiple types of exceptions, use separate except blocks for each exception.
Use finally block for cleanup: If you have cleanup or resource release code that needs to be executed regardless of whether an exception occurred,
place it in a finally block. 
Handle exceptions at the right level: Place your exception handling code at the appropriate level in your code hierarchy.
Handle exceptions where you can effectively recover or take appropriate actions.
Avoid excessive nesting: Excessive nesting of try-except blocks can make the code harder to read and maintain. 
Try to keep the exception handling code as flat as possible by refactoring and using functions or context managers.
Use specific exception messages: When raising exceptions, provide meaningful and descriptive error messages. 
Use custom exception classes: In addition to built-in exception classes, you can create your own custom exception classes
to represent specific types of errors in your code.
Log exceptions: Logging exceptions can be helpful for troubleshooting and debugging. Use a logging framework to log 
exceptions along with relevant information such as the stack trace, timestamp, and any other relevant details.
Don't ignore exceptions: Avoid silently ignoring exceptions without taking any action. If you catch an exception and
decide not to handle it, at least log the exception or provide some form of feedback to indicate that an error occurred.
Test exception handling: Write test cases that cover various scenarios, including exceptions, to ensure that your
exception handling code functions as expected.
```
